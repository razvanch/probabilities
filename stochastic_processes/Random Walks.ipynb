```python
import typing
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


def integer_walk(n: int):
    return pd.Series(
        2 * np.random.randint(0, 2, size=n) - 1
    ).cumsum()


def gaussian_walk(n: int):
    return pd.Series(np.random.normal(size=n)).cumsum()


def plot_walks(walk_f: typing.Callable, n_steps: int, n_walks: int=10):
    df = pd.DataFrame({
        f'walk_{i}': walk_f(n_steps)
        for i in range(10)
    })
    df['sqrt'] = np.sqrt(np.linspace(0, n_steps, n_steps))
    df['-sqrt'] = -df['sqrt']
    # Don't plot too many points
    plot_freq = max(1, n_steps // 10000)
    plot = df.iloc[::plot_freq].plot(figsize=(16, 9))
    plt.title(f'{walk_f.__name__}, {n_steps} steps')
    plt.show()
    
    
    
plot_walks(integer_walk, n_steps=100000)
plot_walks(gaussian_walk, n_steps=100000)
```
